### Convert a PUBMD File to CSV

```python
import pandas as pd
```

```python
filename = "pubmed-virus-set_10000.txt"
SOURCE = f'../Src/{filename}'
DESTINATION = f"../Src_Parsed/{filename.split('.')[0]}.csv"
```

```python
text_file = open(SOURCE,'r',encoding='utf8')
```

```python
# Split Items
text_file.seek(0)
previous_key = None
rows = [{}]

for line in text_file.readlines():
    if(line.startswith("\n")):
        is_append = True
        rows.append({})
        continue
    
    if(line.startswith(" ")):
        rows[-1][previous_key] = rows[-1][previous_key] + line.strip().replace("\n","")
        continue
    
    previous_key,content = line.split("-",1)
    previous_key = previous_key.strip()
    rows[-1][previous_key] = content.strip()
```

```python
pd.DataFrame(rows).to_csv(DESTINATION,index=False)
```
